```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

Function to decide pixel values.

```python


def pixelVal(pix, r1, s1, r2, s2):
	if (0 <= pix and pix <= r1):
		return (s1 / r1)*pix
	elif (r1 < pix and pix <= r2):
		return ((s2 - s1)/(r2 - r1)) * (pix - r1) + s1
	else:
		return ((255 - s2)/(255 - r2)) * (pix - r2) + s2
```

Loading the image

```python

img = cv2.imread(r'../Pics/camera.jpg')
plt.imshow(img)
```

Setting different parameter values.

```python
r1,s1,r2,s2 = 70,0,140,255
```

```python
r1,s1,r2,s2 = 60,0,150,255
```

```python
r1,s1,r2,s2 = 10,0,190,255
```

Applying vectorization.

```python
pixelVal_vec = np.vectorize(pixelVal)
```

Contrast Stretching

```python



contrast_stretched = pixelVal_vec(img, r1, s1, r2, s2)
plt.imshow(contrast_stretched)


```

Save the transformed image.

```python
cv2.imwrite(r'../Temp/contrast_stretch_'+str(r1)+'_'+str(s1)+'_'+str(r2)+'_'+str(s2)+'.jpg', contrast_stretched)
```
