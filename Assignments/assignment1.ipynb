### Assignment 1

Consider any image with size 1024x1024. Modify the image to the sizes 512x512, 256x256, 128x128, 64x64 and 32x32 using subsampling technique. Create the original image from all the above subsampled images using resampling technique. Read any image. Display the histogram, Equalized histogram, and image with equalized histogram


```python
import numpy as np
from matplotlib import pyplot as plt
from skimage.io import imread, imshow , imsave
```

```python
FILEPATH = r"../Pics/nike.jpg"
```

```python
im = imread(FILEPATH)
print(im.shape)
im
```

```python
im = imread(FILEPATH)
print(im.shape)
plt.figure()
imshow(im)
plt.show()
```

```python
def subsample(image, factor: int):
    return image[::factor , ::factor]
```

Subsampling image by a factor of 2.
This will convert the 1024x1024 image to 512x512 image.

```python
image2 = subsample(im , 2)
```

```python
print(image2.shape)
plt.figure()
imshow(image2)
plt.show()
```

Subsampling image by a factor of 4.
This will convert the 1024x1024 image to 256x256 image.

```python
factor = 4
image2 = subsample(im , factor)
print(image2.shape)
plt.figure()
imshow(image2)
plt.show()
```

Subsampling image by a factor of 8.
This will convert the 1024x1024 image to 128x128 image.

```python
factor = 8
image2 = subsample(im , factor)
print(image2.shape)
plt.figure()
imshow(image2)
plt.show()
```

Subsampling image by a factor of 16.
This will convert the 1024x1024 image to 64x64 image.

```python
factor = 16
image2 = subsample(im , factor)
print(image2.shape)
plt.figure()
imshow(image2)
plt.show()
```

Subsampling image by a factor of 32.
This will convert the 1024x1024 image to 32x32 image.

```python
factor = 32
image2 = subsample(im , factor)
print(image2.shape)
plt.figure()
imshow(image2)
plt.show()

# plt.savefig(image2)
imsave(r"output32x32.jpg",image2,check_contrast=False)

```
