### Assignment 2

Consider any image with size 1024x1024. Modify the image to the sizes 512x512, 256x256, 128x128, 64x64 and 32x32 using subsampling technique. Create the original image from all the above subsampled images using resampling technique.


```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

Loading the image

```python
image_file_path = r"../Pics/nike.jpg"
image = cv2.imread(image_file_path)
print(image.shape)

print(f"ASPECT RATIO = {round(image.shape[1] / image.shape[0],2)}")
```

```python
plt.imshow(image)
plt.show()
```

Subsampling the image.

```python
new_size = (64,64)
subsampled_image = cv2.resize(image ,new_size )

print(subsampled_image.shape)
plt.imshow(subsampled_image)
cv2.imwrite("../Temp/subsample64x64.jpg" , subsampled_image)

plt.show()
```

Resampling the image

```python
new_size = (1024,1024)
image_file_path = r"../Temp/subsample64x64.jpg"

image = cv2.imread(image_file_path)
rescaled_image = cv2.resize(image ,new_size )

print(rescaled_image.shape)
plt.imshow(rescaled_image)
cv2.imwrite("../Temp/resample64x64.jpg" , subsampled_image)

plt.show()
```
