```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

Loading the image.

```python
image_file_path = r"../Pics/nissangtr.jpg"
image = cv2.imread(image_file_path)
print(image.shape)

print(f"ASPECT RATIO = {round(image.shape[1] / image.shape[0],2)}")
```

```python
plt.imshow(image)
plt.show()
```

Subsampling the image.

```python
new_size = (128,128)
subsampled_image = cv2.resize(image ,new_size )

print(subsampled_image.shape)
plt.imshow(subsampled_image)
cv2.imwrite("../Temp/subsample128x128.jpg" , subsampled_image)
plt.show()
```

```python
Resampling the image.
```

```python
new_size = (1024,1024)
image_file_path = r"../Temp/subsample128x128.jpg"
save_file_path = r"../Temp/rescaled128x128.jpg"
image = cv2.imread(image_file_path)
rescaled_image = cv2.resize(image ,new_size )

print(rescaled_image.shape)
plt.imshow(rescaled_image)
cv2.imwrite(save_file_path , rescaled_image)
plt.show()
```
