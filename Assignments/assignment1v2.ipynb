Consider any image with size 1024x1024. Modify the image to the sizes 512x512, 256x256, 128x128, 64x64 and 32x32 using subsampling technique. 1. Loading the Modules

```python
import cv2
from matplotlib import pyplot as plt
```

2. Reading an image of resolution 1024x1024 

```python
image = cv2.imread(r"../Pics/nike.jpg")
print(image.shape)
```

```python
plt.imshow(image)
plt.show()
```

Subsampling Image to 512x512 , 256x256 , 128x128 , 64x64 , 32x32 Images

```python
new_size = (512,512)
subsampled_image = cv2.resize(image ,new_size )

print(subsampled_image.shape)
plt.imshow(subsampled_image)

cv2.imwrite("../Temp/subsample512x512.jpg" , subsampled_image)
plt.show()
```
